```python
import os
import pandas as pd
import matplotlib.pyplot as plt
```

```python
all_files = [file for file in os.listdir('./Sales_Data')]
all_data = pd.DataFrame()
for file in all_files:
    df = pd.read_csv('./Sales_Data/' + file)
    all_data = pd.concat([all_data,df])
```

```python
all_data.to_csv('all_data.csv',index=False)
```

```python
raw_data = pd.read_csv('./all_data.csv')
raw_data.head()
```

```python
raw_data['Month'] = raw_data['Order Date'].str[0:2]
raw_data = raw_data.dropna(how='all')
raw_data
```

```python
raw_data = raw_data[raw_data['Order Date'].str[0:2] != 'Or']
raw_data['Quantity Ordered'] = pd.to_numeric(raw_data['Quantity Ordered'])
raw_data['Price Each'] = pd.to_numeric(raw_data['Price Each'])
raw_data['Month'] = raw_data['Month'].astype('int32')
raw_data
```

```python
raw_data["Sales"] = raw_data['Quantity Ordered'] * raw_data['Price Each']
raw_data
```

```python
results = raw_data.groupby('Month').sum()
```

```python
months = range(1,13)
plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('Sales ($)')
plt.xlabel('Months')
plt.show
```

```python
def get_city(address):
    return address.split(',')[1]

def get_city_code(address):
    return address.split(',')[2].split(' ')[1]

raw_data['City'] = raw_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_city_code(x)})")
raw_data
```

```python
city_sales = raw_data.groupby('City').sum()
city_sales
```

```python
city = raw_data['City'].unique()
city = [city for city,df in raw_data.groupby('City')]
plt.bar(city, city_sales['Sales'])
plt.xticks(city,rotation='vertical',size=8)
plt.ylabel('Sales ($)')
plt.xlabel('Cities')
plt.show()
```

```python
raw_data['Order Date'] =  pd.to_datetime(raw_data['Order Date'])
raw_data
```

```python
raw_data['Hour'] = raw_data['Order Date'].dt.hour
raw_data['Minute'] = raw_data['Order Date'].dt.minute
raw_data['Count'] = 1
raw_data
```

```python
hours = [hour for hour, df in raw_data.groupby('Hour')]
plt.plot(hours,raw_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.grid()
plt.show()
```

```python
df = raw_data[raw_data['Order ID'].duplicated(keep=False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df2 = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
df2
```

```python
from itertools import combinations
from collections import Counter

count = Counter()

for row in df2['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
for key,value in count.most_common(10):
    print(key,value)
```

```python
product_group = raw_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]

plt.bar(products,quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Products')
plt.xticks(products,rotation='vertical',size=8)
plt.show()
```

```python
prices = raw_data.groupby('Product').mean()['Price Each']
prices
```

```python
fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color='g')
ax2.plot(products, prices, color='b')

ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
